```python

import os
import random
import math
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt


SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)

def generate_multivariate_series(n_steps=2000, n_vars=3, freq=24, noise_std=0.3):

    t = np.arange(n_steps)
    df = pd.DataFrame()
    # base seasonality and trend for var0
    trend0 = 0.001 * t
    season0 = 2.0 * np.sin(2 * np.pi * t / freq + 0.1)
    var0 = 3.0 + trend0 + season0 + np.random.normal(scale=noise_std, size=n_steps)
    df['var0'] = var0

    # var1 depends on var0 nonlinearly + its own seasonality
    season1 = 1.5 * np.cos(2 * np.pi * t / (freq*1.5) + 0.4)
    var1 = 1.0 + 0.5 * np.tanh(var0) + 0.5 * season1 + np.random.normal(scale=noise_std, size=n_steps)
    df['var1'] = var1

    # var2 with multiplicative interaction and occasional jumps
    season2 = 0.8 * np.sin(2 * np.pi * t / (freq*0.75) - 0.2)
    var2 = 0.5 + 0.3 * var0 * (1 + 0.1 * np.sin(t/50.0)) + season2 + np.random.normal(scale=noise_std, size=n_steps)
    df['var2'] = var2

    # Add occasional jumps
    for j in range(50, n_steps, 400):
        df.iloc[j:j+5] += np.random.normal(loc=1.0, scale=0.5, size=(5, n_vars))

    return df


def make_windows(data, input_len=48, forecast_horizon=1, step=1):

    T, n_vars = data.shape
    X, y = [], []
    for start in range(0, T - input_len - forecast_horizon + 1, step):
        end = start + input_len
        X.append(data[start:end])
        y.append(data[end:end+forecast_horizon, 0])  # predict var0 by default
    X = np.array(X)
    y = np.array(y)
    if forecast_horizon == 1:
        y = y.reshape(-1, 1)
    return X, y

-
class AttentionLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        # input_shape: (batch, time, features)
        self.W = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                 initializer='glorot_uniform',
                                 trainable=True, name='W_att')
        self.b = self.add_weight(shape=(input_shape[-1],),
                                 initializer='zeros', trainable=True, name='b_att')
        self.u = self.add_weight(shape=(input_shape[-1],),
                                 initializer='glorot_uniform',
                                 trainable=True, name='u_att')

    def call(self, inputs, mask=None):
        # inputs: (batch, time, features)
        uit = tf.tanh(tf.tensordot(inputs, self.W, axes=1) + self.b)  # (batch, time, features)
        ait = tf.tensordot(uit, self.u, axes=1)  # (batch, time)
        a = tf.nn.softmax(ait, axis=1)  # (batch, time)
        a = tf.expand_dims(a, axis=-1)  # (batch, time, 1)
        weighted_input = inputs * a  # (batch, time, features)
        output = tf.reduce_sum(weighted_input, axis=1)  # (batch, features)
        return output

    def get_config(self):
        base = super(AttentionLayer, self).get_config()
        return base


def transformer_encoder_block(dim, num_heads, ff_dim, dropout=0.1, name=None):
    input_layer = layers.Input(shape=(None, dim))
    # Multi-Head Self Attention
    x = layers.MultiHeadAttention(num_heads=num_heads, key_dim=dim//num_heads, name=f'{name}_mha')(input_layer, input_layer)
    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(input_layer + x)
    # Feed-forward
    ff = layers.Dense(ff_dim, activation='relu')(x)
    ff = layers.Dense(dim)(ff)
    ff = layers.Dropout(dropout)(ff)
    out = layers.LayerNormalization(epsilon=1e-6)(x + ff)
    return keras.Model(inputs=input_layer, outputs=out, name=name)


def build_lstm_baseline(input_shape, units=64, dropout=0.1):
    inp = layers.Input(shape=input_shape)
    x = layers.LSTM(units, return_sequences=False)(inp)
    x = layers.Dropout(dropout)(x)
    out = layers.Dense(1)(x)  # predict scalar (var0)
    model = keras.Model(inp, out, name='lstm_baseline')
    model.compile(optimizer='adam', loss='mse', metrics=[keras.metrics.MeanAbsoluteError()])
    return model

def build_lstm_attention(input_shape, units=64, dropout=0.1):
    inp = layers.Input(shape=input_shape)
    x = layers.Bidirectional(layers.LSTM(units, return_sequences=True))(inp)
    # attention
    att = AttentionLayer()(x)  # (batch, features)
    x = layers.Dropout(dropout)(att)
    out = layers.Dense(1)(x)
    model = keras.Model(inp, out, name='lstm_attention')
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=3e-4), loss='mse', metrics=[keras.metrics.MeanAbsoluteError()])
    return model

def build_transformer(input_shape, dim=64, num_heads=4, ff_dim=128, num_layers=2, dropout=0.1):

    timesteps, features = input_shape
    inp = layers.Input(shape=input_shape)
    # project features
    x = layers.Dense(dim)(inp)
    # positional encoding (learned)
    pos = layers.Embedding(input_dim=timesteps+1, output_dim=dim)
    positions = tf.range(start=0, limit=timesteps, delta=1)
    pos_encoding = pos(positions)
    x = x + pos_encoding

    # stack transformer encoder blocks
    for i in range(num_layers):
        block = transformer_encoder_block(dim, num_heads, ff_dim, dropout, name=f'transformer_block_{i}')
        x = block(x)


    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(dropout)(x)
    out = layers.Dense(1)(x)  # forecast scalar
    model = keras.Model(inp, out, name='transformer')
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=3e-4), loss='mse', metrics=[keras.metrics.MeanAbsoluteError()])
    return model

def directional_accuracy(y_true, y_pred):

    raise NotImplementedError("Use directional_accuracy_vs_prev to supply prev_actuals")

def directional_accuracy_vs_prev(y_true, y_pred, prev_actuals):

    true_dir = np.sign(y_true.flatten() - prev_actuals.flatten())
    pred_dir = np.sign(y_pred.flatten() - prev_actuals.flatten())
    # treat zeros as no-change; count matches
    match = (true_dir == pred_dir)
    return np.mean(match)


def evaluate_model(model, X_test, y_test, scaler_y=None, X_raw_prev=None):

    y_pred = model.predict(X_test, batch_size=64)
    if scaler_y is not None:
        y_pred = scaler_y.inverse_transform(y_pred.reshape(-1, 1)).reshape(-1, 1)
        y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1)).reshape(-1, 1)
    else:
        y_true = y_test

    mae = mean_absolute_error(y_true, y_pred)
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    if X_raw_prev is not None:
        # X_raw_prev is last step value of the target variable in the input window
        da = directional_accuracy_vs_prev(y_true, y_pred, X_raw_prev.reshape(-1, 1))
    else:
        da = np.nan
    return {'mae': mae, 'rmse': rmse, 'directional_accuracy': da, 'y_true': y_true, 'y_pred': y_pred}


def main():
    # generate data
    n_steps = 3000
    df = generate_multivariate_series(n_steps=n_steps, n_vars=3, freq=24, noise_std=0.25)
    print("Generated data shape:", df.shape)
    df.plot(subplots=True, figsize=(10,6))
    plt.suptitle("Synthetic Multivariate Time Series (var0 target)")
    plt.show()

    arr = df.values  # shape (T, n_vars)

    INPUT_LEN = 48
    HORIZON = 1
    X, y = make_windows(arr, input_len=INPUT_LEN, forecast_horizon=HORIZON, step=1)
    print("X shape, y shape:", X.shape, y.shape)


    prev_vals = X[:, -1, 0].reshape(-1, 1)


    N = X.shape[0]
    train_end = int(N * 0.7)
    val_end = int(N * 0.85)

    X_train, y_train = X[:train_end], y[:train_end]
    X_val, y_val = X[train_end:val_end], y[train_end:val_end]
    X_test, y_test = X[val_end:], y[val_end:]
    prev_train = prev_vals[:train_end]
    prev_val = prev_vals[train_end:val_end]
    prev_test = prev_vals[val_end:]

    print("Train/Val/Test sizes:", X_train.shape[0], X_val.shape[0], X_test.shape[0])

    n_vars = X.shape[-1]
    scalers_X = []
    X_train_reshaped = X_train.reshape(-1, n_vars)
    for i in range(n_vars):
        sc = StandardScaler()
        X_train_reshaped[:, i:i+1] = sc.fit_transform(X_train_reshaped[:, i:i+1])
        scalers_X.append(sc)
    X_train_scaled = X_train_reshaped.reshape(X_train.shape)


    def scale_X(X_in, scalers):
        Xr = X_in.reshape(-1, n_vars)
        for i, sc in enumerate(scalers):
            Xr[:, i:i+1] = sc.transform(Xr[:, i:i+1])
        return Xr.reshape(X_in.shape)

    X_val_scaled = scale_X(X_val, scalers_X)
    X_test_scaled = scale_X(X_test, scalers_X)


    scaler_y = StandardScaler()
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1))
    y_val_scaled = scaler_y.transform(y_val.reshape(-1,1))
    y_test_scaled = scaler_y.transform(y_test.reshape(-1,1))


    input_shape = (INPUT_LEN, n_vars)
    lstm = build_lstm_baseline(input_shape, units=64, dropout=0.2)
    lstm_att = build_lstm_attention(input_shape, units=64, dropout=0.2)
    transformer = build_transformer(input_shape, dim=64, num_heads=4, ff_dim=128, num_layers=2, dropout=0.1)


    lstm.summary()
    lstm_att.summary()
    transformer.summary()

    # callbacks
    early = keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
    chk_dir = 'saved_models'
    os.makedirs(chk_dir, exist_ok=True)

    # train baseline LSTM
    print("\nTraining baseline LSTM (no attention)...")
    history_lstm = lstm.fit(X_train_scaled, y_train_scaled,
                            validation_data=(X_val_scaled, y_val_scaled),
                            epochs=60, batch_size=64, callbacks=[early], verbose=2)
    lstm.save(os.path.join(chk_dir, 'lstm_baseline.h5'))

    # train LSTM + Attention
    print("\nTraining LSTM + Attention...")
    history_lstm_att = lstm_att.fit(X_train_scaled, y_train_scaled,
                                    validation_data=(X_val_scaled, y_val_scaled),
                                    epochs=80, batch_size=64, callbacks=[early], verbose=2)
    lstm_att.save(os.path.join(chk_dir, 'lstm_attention.h5'))

    # train Transformer
    print("\nTraining Transformer encoder...")
    history_trans = transformer.fit(X_train_scaled, y_train_scaled,
                                    validation_data=(X_val_scaled, y_val_scaled),
                                    epochs=80, batch_size=64, callbacks=[early], verbose=2)
    transformer.save(os.path.join(chk_dir, 'transformer.h5'))

    # Evaluate
    print("\nEvaluating on test set (inverse scaling applied)...")
    results = {}
    results['LSTM'] = evaluate_model(lstm, X_test_scaled, y_test_scaled, scaler_y=scaler_y, X_raw_prev=prev_test)
    results['LSTM_ATT'] = evaluate_model(lstm_att, X_test_scaled, y_test_scaled, scaler_y=scaler_y, X_raw_prev=prev_test)
    results['Transformer'] = evaluate_model(transformer, X_test_scaled, y_test_scaled, scaler_y=scaler_y, X_raw_prev=prev_test)


    comp = []
    for k, v in results.items():
        comp.append({
            'model': k,
            'mae': v['mae'],
            'rmse': v['rmse'],
            'directional_accuracy': v['directional_accuracy']
        })
    comp_df = pd.DataFrame(comp).sort_values('mae')
    print("\nModel comparison on test set (lower MAE better):")
    print(comp_df.to_string(index=False))

    # Plot predictions vs actual for best model
    best_model_key = comp_df.iloc[0]['model']
    best = results[best_model_key]
    y_true = best['y_true'].flatten()
    y_pred = best['y_pred'].flatten()
    plt.figure(figsize=(12,5))
    plt.plot(y_true, label='actual', alpha=0.8)
    plt.plot(y_pred, label='predicted', alpha=0.8)
    plt.title(f"Test set predictions - Best model: {best_model_key}")
    plt.legend()
    plt.show()

    # Save a CSV comparing predictions
    pd.DataFrame({'y_true': y_true, 'y_pred': y_pred}).to_csv(os.path.join(chk_dir, f'preds_{best_model_key}.csv'), index=False)
    print(f"\nSaved best model predictions to {os.path.join(chk_dir, f'preds_{best_model_key}.csv')}")

    # Simple hyperparameter grid (example) - small for speed
    print("\nRunning small grid search example (LSTM units: [32,64]) - this is illustrative and kept small.")
    grid_results = []
    for units in [32, 64]:
        model_tmp = build_lstm_attention(input_shape, units=units, dropout=0.2)
        print(f"Training LSTM+Att units={units}")
        model_tmp.fit(X_train_scaled, y_train_scaled, validation_data=(X_val_scaled, y_val_scaled), epochs=20, batch_size=64, callbacks=[early], verbose=0)
        res = evaluate_model(model_tmp, X_test_scaled, y_test_scaled, scaler_y=scaler_y, X_raw_prev=prev_test)
        grid_results.append({'units': units, 'mae': res['mae'], 'rmse': res['rmse']})
        tf.keras.backend.clear_session()
    print(pd.DataFrame(grid_results))

if __name__ == "__main__":
    main()
```
